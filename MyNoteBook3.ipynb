```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
data=pd.read_csv("D:/Ml/datasets/chd_data.csv")
predictors=["age"]
target="chd"
x=data[predictors]
y=data[target]
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=0)
model=LogisticRegression()
print("Shape of x_train:",xtrain.shape)
print("Shape of y_train:",ytrain.shape)
print("Shape of x_test:",xtest.shape)
print("Shape of y_test:",ytest.shape)
model.fit(xtrain,ytrain)
print("Intercept:",model.intercept_,"\nCoefficients:",model.coef_)
test=np.array([30]).reshape(1,-1)
print("Predicted probability:",model.predict_proba(test))
print("\nPredicted target:",model.predict(test))
fig = plt.figure(figsize = (10, 5))
plt.bar(xtrain[predictors].iloc[:20, 0], ytrain.iloc[:20], color="darkblue", width=0.4)
plt.xlabel("Age")
plt.ylabel("Coronary_heart_disease")
plt.title("Age Vs Coronary Heart Disease in Bar Chart")
plt.show()
plt.scatter(xtrain[predictors].iloc[:20,0],ytrain.iloc[:20],color="red",marker="o")
plt.xlabel("Age")
plt.ylabel("Coronary_heart_disease")
plt.title("Age Vs Coronary Heart Disease as Scattered Plot")
plt.show()
plt.hist(data, bins=5, color=["yellow","green"])
plt.xlabel("Age")
plt.ylabel("Coronary_heart_disease")
plt.title("Age Vs Coronary Heart Disease in Histogram")
plt.show()


```
